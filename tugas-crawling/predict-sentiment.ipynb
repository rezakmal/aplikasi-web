```python
import pickle

feature_bow = pickle.load(open('feature-bow.p', 'rb'))
model_nb = pickle.load(open('model-nb.p', 'rb'))
model_nn = pickle.load(open('model-nn.p', 'rb'))
```

```python
#predict sentiment dari kumpulan kalimat

import pandas as pd

data = pd.read_csv('data/wasit_bahrain.csv', sep=',')
```

```python
data.head()
```

```python
data = data.rename(columns={'full_text': 'tweet'})
```

```python
data.head()
```

```python
import re

def cleansing(sent):
    # Mengubah semua huruf menjadi huruf kecil
    string = sent.lower()
    
    # Menghapus URL
    string = re.sub(r'http\S+|www\S+|https\S+', '', string, flags=re.MULTILINE)
    
    # Menghapus mention (@username)
    string = re.sub(r'@\w+', '', string)
    
    # Menghapus hashtag (tetap menyimpan kata tanpa tanda '#')
    string = re.sub(r'#(\w+)', r'\1', string)
    
    # Menghapus angka dan tanda baca, hanya menyisakan huruf
    string = re.sub(r'[^a-zA-Z\s]', ' ', string)
    
    # Menghapus RT
    string = re.sub(r'\brt\b', '', string)
    
    # Menghapus whitespace berlebih
    string = re.sub(r'\s+', ' ', string).strip()
    
    return string
```

```python
def predict_sentiment(sent):
    # cleansing
    text = cleansing(str(sent))
    # feature extraction
    text_feature = feature_bow.transform([text])
    # predict
    return model_nb.predict(text_feature)[0]
```

```python
data['tweet'] = data.tweet.apply(cleansing)
data.head()
```

```python
data['predicted_sentiment'] = data.tweet.apply(predict_sentiment)
```
